```python
import pandas as pd
```

```python
citi_travel1912 = pd.read_csv('201912-citibike-tripdata.csv')
```

```python
citi_travel1912.head()
```

```python
unknown = citi_travel1912.loc[(citi_travel1912['gender'] == 0)]
```

```python
clean_df = citi_travel1912.drop(unknown.index)
```

```python
clean_df
```

```python
trip_df = clean_df[['tripduration', 'end station name']].groupby(by = "end station name").count()
trip_df.head()
```

```python
trip_df.rename(columns={'tripduration':'trips'}, inplace = True)
```

```python
trip_df
```

```python
trip_df.to_csv('cleandata/201912citibiketrips.csv')
```

```python
customer_df = clean_df.loc[(clean_df['usertype'] == 'Customer')]
```

```python
customer_df.to_csv('cleandata/201912customerdata.csv')
```
