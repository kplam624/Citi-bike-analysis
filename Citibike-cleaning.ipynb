```python
# import the dependencies.
import pandas as pd
```

```python
# Reads the csv files.
citi_travel1912 = pd.read_csv('201912-citibike-tripdata.csv')
citi_travel1512 = pd.read_csv('201512-citibike-tripdata.csv')
citi_travel2012 = pd.read_csv('202012-citibike-tripdata.csv')

# Shows the first dataframe
citi_travel1912.head()
```

```python
# We want to remove the unknown values.
# First by locating the unknown genders
unknown1 = citi_travel1912.loc[(citi_travel1912['gender'] == 0)]
unknown2 = citi_travel1512.loc[(citi_travel1512['gender'] == 0)]
unknown3 = citi_travel2012.loc[(citi_travel2012['gender'] == 0)]

# Removing the rows.
clean_df1 = citi_travel1912.drop(unknown1.index)
clean_df2 = citi_travel1512.drop(unknown2.index)
clean_df3 = citi_travel2012.drop(unknown3.index)

clean_df1.head()
```

```python
# Creating csv files to calculate the trips taken.
trip_df1 = clean_df1[['tripduration', 'end station name']].groupby(by = "end station name").count()
trip_df2 = clean_df2[['tripduration', 'end station name']].groupby(by = "end station name").count()
trip_df3 = clean_df3[['tripduration', 'end station name']].groupby(by = "end station name").count()
trip_df1.head()
```

```python
trip_df3.head()
```

```python
# For better readability
trip_df1.rename(columns={'tripduration':'trips'},inplace = True)
trip_df2.rename(columns={'tripduration':'trips'},inplace = True)
trip_df3.rename(columns={'tripduration':'trips'},inplace = True)

# Shows the modified table.
trip_df1.head()
```

```python
# Creating the csvs
trip_df1.to_csv('cleandata/201912citibiketrips.csv')
trip_df2.to_csv('cleandata/201512citibiketrips.csv')
trip_df3.to_csv('cleandata/202012citibiketrips.csv')
```

```python
clean_df1 = clean_df1.replace({'gender': {1: 'Male', 2: 'Female'}})
clean_df2 = clean_df2.replace({'gender': {1: 'Male', 2: 'Female'}})
clean_df3 = clean_df3.replace({'gender': {1: 'Male', 2: 'Female'}})
```

```python
clean_df1.head()
```

```python
# Focuses on the customers not subscribers
customer_df1 = clean_df1.loc[(clean_df1['usertype'] == 'Customer')]
customer_df1.to_csv('cleandata/201912customerdata.csv', index = False)

customer_df3 = clean_df3.loc[(clean_df3['usertype'] == 'Customer')]
customer_df3.to_csv('cleandata/202012customerdata.csv', index = False)

customer_df1.head()
```

```python
# Create a file for subscribers in 2019.
subscriber_df1 = clean_df1.loc[(clean_df1['usertype'] == 'Subscriber')]
subscriber_df1.to_csv('cleandata/201912subscriberdata.csv', index = False)

# Create a file for subscribers in 2015. These were the only user types that year.
subscriber_df2 = clean_df2.loc[(clean_df2['usertype'] == 'Subscriber')]
subscriber_df2.to_csv('cleandata/201512subscriberdata.csv', index = False)

subscriber_df1
```

```python
# Creates a csv with only locations.
station_location_df = clean_df3[['end station name', 'end station latitude', 'end station longitude']].drop_duplicates()
station_location_df.to_csv('cleandata/end_station_location.csv', index = False)
```
